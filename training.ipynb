```python
import tensorflow as tf
```

```python


# define the input layer with 121 neurons
input_layer = tf.keras.layers.Input(shape=(121,))

# define the first hidden layer with 60 neurons and ReLU activation function
hidden_layer_1 = tf.keras.layers.Dense(60, activation='relu')(input_layer)

# define the second hidden layer with 60 neurons and ReLU activation function
hidden_layer_2 = tf.keras.layers.Dense(60, activation='relu')(hidden_layer_1)

# define the output layer with 3 neurons and sigmoid activation function
output_layer = tf.keras.layers.Dense(3, activation='sigmoid')(hidden_layer_2)

# create the model with input and output layers
model = tf.keras.models.Model(inputs=input_layer, outputs=output_layer)

# compile the model with MSE (mean squared error) loss function and Adam optimizer
model.compile(loss='mse', optimizer='adam')
```

## Training

```python
for i in range(5):
  df = pd.read_csv(f"C:\\Users\\shrut\\OneDrive\\Desktop\\miniproject\\code(20.3.23)\\barn{i}.csv")
  lst = [i for i in range(0,121)]
  X = df.iloc[:, lst].values
  y = df.iloc[:, [123,124,125]].values
  X_train = np.asarray(X).astype('float32')
  y_train = np.asarray(y).astype('float32')
  #X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.3)
  
  # train the model using backpropagation
  model.fit(X_train, y_train, epochs=500, batch_size=400)
  # evaluate the model on test data
  loss = model.evaluate(X_train, y_train,verbose=0)
  # print the accuracy
  print('Loss:', loss)
```
